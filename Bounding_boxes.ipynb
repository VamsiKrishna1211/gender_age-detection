```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np
```

```python
image_data = pd.read_csv("train_faces_concactenated_data_with_path_with_age_range.csv")
```

```python
image_data.columns
```

```python
x = image_data['x']
y = image_data['y']
dx = image_data['dx']
dy = image_data['dy']
```

```python
index = 50
im = np.array(Image.open(image_data['path'][index]), dtype=np.uint8)
fig,ax = plt.subplots(1)
ax.imshow(im)
rect = patches.Rectangle((x[index],y[index]),dx[index],dy[index],linewidth=10,edgecolor='r',facecolor='none')
ax.add_patch(rect)
plt.show()
```

```python
rect = patches.Rectangle((x[index],y[index]),dx[index],dy[index],linewidth=1,edgecolor='r',facecolor='none')
ax.add_patch(rect)
plt.show()
```
