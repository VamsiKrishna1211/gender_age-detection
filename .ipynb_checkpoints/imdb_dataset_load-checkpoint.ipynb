```python
import scipy.io 
import numpy as np
import cv2
import scipy.io
import argparse
from tqdm import tqdm
from utils import get_meta
import utils
import pandas as pd
```

```python
def main():
    #args = get_args()
    #output_path = 
    db = 'imdb'
    img_size = 257
    min_score = 1.0
    root_path = "./Dataset-copy/imdb_faces/imdb_crop/"
    
    path = []
    mat_path = root_path + "{}.mat".format(db)
    full_path, dob, gender, photo_taken, face_score, second_face_score, age = get_meta(mat_path, db)
    out_genders = []
    out_ages = []
    face_scores = []
    sample_num = len(face_score)
    #out_imgs = np.empty((sample_num, img_size, img_size, 3), dtype=np.uint8)
    valid_sample_num = 0
    #return full_path

    for i in tqdm(range(sample_num)):
        if face_score[i] < min_score:
            continue

        if (~np.isnan(second_face_score[i])) and second_face_score[i] > 0.0:
            continue

        if ~(0 <= age[i] <= 100):
            continue

        if np.isnan(gender[i]):
            continue

        out_genders.append(int(gender[i]))
        out_ages.append(age[i])
        #img = cv2.imread(root_path + str(full_path[i][0]))
        image_path = root_path + str(full_path[i][0])
        #out_imgs[valid_sample_num] = img#cv2.resize(img, (img_size, img_size))
        valid_sample_num += 1
        path.append(image_path)
        face_scores.append(face_score[i])
        
        
        
    print("done creating DB")
    #output = {"image_path": image_paths[:valid_sample_num], "gender": np.array(out_genders), "age": np.array(out_ages),
              #"db": db, "img_size": img_size, "min_score": min_score}
    imdb_processed_data = np.vstack((path,out_ages,out_genders,face_scores)).T
    return imdb_processed_data
    
```

```python
#if __name__ == '__main__':
full_path = main()
```
